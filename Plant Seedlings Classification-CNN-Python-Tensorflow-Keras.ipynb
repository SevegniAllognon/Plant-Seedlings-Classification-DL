# Building dataset from images

```python
train_ds_ = tf.keras.preprocessing.image_dataset_from_directory(
    data_direc_train,
    label_mode='categorical',
    validation_split=0.8, #80% training
    subset="training",
    seed=123,
    image_size=(img_height, img_width),
    batch_size=batch_size)
```

# Data augmentation

```python
data_augmentation = keras.Sequential(
    [
        layers.experimental.preprocessing.RandomFlip("horizontal", 
                                    input_shape=(img_height, img_width,3)),
        layers.experimental.preprocessing.RandomRotation(0.1),
        layers.experimental.preprocessing.RandomZoom(0.1),
    ]
)
```

# Building CNN Model

```python
def cnn_model():
    
    model = tf.keras.models.Sequential()
    model.add(tf.keras.layers.InputLayer(input_shape=(img_width,img_height,3,))) 
    model.add(tf.keras.layers.Conv2D(64, kernel_size=(3,3), activation='relu')) 
    model.add(tf.keras.layers.MaxPool2D(pool_size = (2,2)))  
    model.add(tf.keras.layers.BatchNormalization()) 
    
    model.add(tf.keras.layers.Conv2D(64, kernel_size=(3,3), strides = (1,1), activation='relu')) 
    model.add(tf.keras.layers.MaxPool2D(pool_size = (2,2)))  
    model.add(tf.keras.layers.BatchNormalization()) 
    
    model.add(tf.keras.layers.Conv2D(128, kernel_size=(3,3), strides = (1,1), activation='relu')) 
    model.add(tf.keras.layers.MaxPool2D(pool_size = (2,2))) 
    model.add(tf.keras.layers.BatchNormalization()) 
    
    model.add(tf.keras.layers.Conv2D(128, kernel_size=(3,3), strides = (1,1), activation='relu')) 
    model.add(tf.keras.layers.MaxPool2D(pool_size = (2,2)))  
    model.add(tf.keras.layers.GlobalMaxPool2D()) 
    
    model.add(tf.keras.layers.Flatten()) 
    model.add(tf.keras.layers.Dense(128, activation='relu'))
    model.add(tf.keras.layers.Dropout(0.2)) 
    model.add(tf.keras.layers.Dense(64, activation='relu'))
    model.add(tf.keras.layers.BatchNormalization()) 
    
    model.add(tf.keras.layers.Dense(12, activation='softmax')) 
    
    return model

epochs = 15
```

```python
model = cnn_model()
model.summary()
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.CategoricalCrossentropy(),
              metrics=['accuracy'])
```

```python
gc.collect()

model_histo = model.fit(train_ds, validation_data=val_ds, epochs=epochs)
```

```python
acc = model_histo.history['accuracy']
val_acc = model_histo.history['val_accuracy']

loss = model_histo.history['loss']
val_loss = model_histo.history['val_loss']

epochs_range = range(epochs)

plt.figure(figsize=(16, 8))

plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')
plt.grid()

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.grid()

plt.show()
```

```python
print("Evaluate train:\n................")
print(model.evaluate(train_ds))  # Evaluate on train set

print("Evaluate validation:\n................")
print(model.evaluate(val_ds))  # Evaluate on validation set
```
